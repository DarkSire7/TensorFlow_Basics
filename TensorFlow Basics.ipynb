```python
import tensorflow as tf
print(tf.version)
```

Write TensorFlow code to create a constant tensor with the values [1, 2, 3].

```python
tensor = tf.constant([1, 2, 3])
print(tensor)
```

Given two tensors a = tf.constant([2, 4, 6]) and b = tf.constant([1, 2, 3]), write TensorFlow code to perform element-wise multiplication.

```python
a = tf.constant([2, 4, 6])
b = tf.constant([1, 2, 3])
c = a*b
print(c)
```

Take a 1D tensor x = tf.constant([1, 2, 3, 4, 5, 6]) and reshape it into a 2D tensor with 2 rows and 3 columns.

```python
x = tf.constant([1, 2, 3, 4, 5, 6])
x = tf.reshape(x, [2, 3])
print(x)
```

Given a 2D tensor matrix = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]]), extract the middle column using TensorFlow.

```python
matrix = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
middle = matrix[:, 1]
print(middle)
```

Initialize a TensorFlow variable weights with a 3x3 matrix of random values.

```python
weights = tf.Variable(tf.random.normal([3,3]))
print(weights)
```

Write TensorFlow code to calculate the element-wise square root of a tensor.

```python
a = tf.constant([4.0, 9.0, 16.0])
b = tf.sqrt(a)
print(b)
```
